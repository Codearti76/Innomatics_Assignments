```python
#Problem 1: Unique Words in a Sentence

# Taking input from user
sentence = input("Enter sentence: ")

# Splitting sentence into words
words = sentence.split()

# Converting list to set to get unique words
uniq_words = set(words)

# Printing results
print("Unique words count:", len(uniq_words))
print("Unique words:", uniq_words)
```

```python
# Problem 2: Highest Salary from Employee Data
# Dictionary with keys as employees and values as salary
employees = {
    "Aarti":71000,
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
# Variable to store the highest value
highest_salary = 0
highest_employee = ""
# Compareing each value till compare every salary
for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name
# Printing Highest salary and the employee name
print("Highest Salary:", highest_employee, "-", highest_salary)
```

```python
# Problem 3: Find Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66, 21]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through list
for num in numbers:
    if num > maximum: # Compare the max value to the num
        maximum = num # Stores num as max if it is greater
    if num < minimum: # Compare the min value to the num
        minimum = num  # Stores num as min if it is smaller

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
#Problem 4: Count Products Above a Price Threshold
# List of the prices
prices = [450, 1200, 899, 1500, 300, 500, 600]
# Counter variable to count the prices > 1000
count = 0
# Loop continues till all the prices are compared
for price in prices:
    if price > 1000:
        count += 1 # Counts only if price > 1000

print("Products above 1000:", count)
```

```python
# Problem 5: Calculate Attendance Percentage
# List of the attendance taken
attendance = ["P", "P", "A", "P", "P", "P", "A", "P", "P", "P"]
# Counter variable to count no. of P's
present_count = 0

for status in attendance:
    if status == "P":
        present_count += 1
# Total no. of days means total no. of entries
total_days = len(attendance)
# Percentage fromula
percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

```python
# Problem 6: Remove Duplicate Phone Numbers
# List of numbers containing duplicates
phone_numbers = [9876543210, 9123456789, 9876543210, 9123456789, 9123456789]

# Convert list to set to remove duplicates
uniq_numbers = set(phone_numbers)

print("Unique phone numbers:", uniq_numbers)
```

```python
# Problem 7: Count Character Frequency
# Input string as per expected output
text = "pythonp"
# Dictionary
frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1 #if the char appears value will increment
    else:
        frequency[char] = 1

print(frequency)
```

```python
# Problem 9: Check if a Key Exists in a Dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000,
    "Aarti": 70000
}

key = "Aarti"

if key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
# Problem 10: Calculate Average Marks
#List of random marks 
marks = [70, 75, 85, 60]

total = 0
# Sums up all the marks
for mark in marks:
    total += mark

average = total / len(marks)

print("Average Marks:", average)
```
