```python
# Problem Statement 1:Smart Parking Lot Management System Design a function to manage a smart parking lot.
def smart_parking(capacity, logs):
    current_vehicles = 0
    peak_usage = 0
    
    for log in logs:
        if log == "IN":
            current_vehicles += 1
        elif log == "OUT":
            if current_vehicles > 0:
                current_vehicles -= 1
        
        # Track peak usage
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles
    
    print("Currently Parked Vehicles:", current_vehicles)
    print("Peak Parking Usage:", peak_usage)
    
    if current_vehicles > capacity:
        print("Alert: Parking Exceeded Capacity!")
    else:
        print("Parking Status: Available")
# Given required input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking(capacity, logs)       
```

```python
#  Problem Statement 2:Onldef estimate_delivery_time(distance, traffic, weather):
    
def estimate_delivery_time(distance, traffic, weather):
    
    # Base time calculation (30 km per hour speed)
    base_time = (distance / 30) * 60  
    
    delay = 0
    
    # Traffic conditions
    if traffic == "Low":
        delay += 5  #as the traffic increases delay increases
    elif traffic == "Medium":
        delay += 10
    elif traffic == "High":
        delay += 20
    
    # Weather conditions
    if weather == "Rainy":
        delay += 10
    elif weather == "Storm":
        delay += 20
    
    final_time = base_time + delay
    
    print("Estimated Delivery Time:", int(final_time), "minutes")
estimate_delivery_time(10, "High", "Rainy")  #function call
```

```python
# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer
def seat_occupancy_analyzer(total_seats, booked_seats):
    
    booked_count = len(booked_seats)
    
    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100
    
    print("Occupancy:", int(occupancy), "%")
    
    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull!!")
    elif occupancy >= 75:
        print("Show Status: Almost Full..")
    else:
        print("Show Status: Seats AvailableðŸ‘Œ")
    
    # Suggest additional show
    if occupancy >= 90:
        print("Suggestion: Consider Opening Additional Show")
total_seats = 200
booked_seats = [1] * 150   # 150 seats booked

seat_occupancy_analyzer(total_seats, booked_seats)
```

```python
# Problem Statement 4:Cloud Server Load Classification System
def classify_server_load(cpu_readings):
    
    total = 0
    
    # Loop to calculate total CPU usage
    for reading in cpu_readings:
        total += reading
    
    # Calculate average
    average = total / len(cpu_readings)
    
    print("Average CPU Load:", int(average), "%")
    
    # Classification
    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: WarningðŸš¨")
    else:
        print("Server Status: Critical")
cpu_data = [45, 60, 70, 85, 90]

classify_server_load(cpu_data)
```

```python
#Problem Statement 5:Smart Classroom Resource Usage Monitor

def classroom_resource_monitor(resource_usage):
    
    overused = []
    
    # Loop through dictionary
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)
    
    # Display overused resources
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(usage_data)
```

```python
# Problem Statement 6: Online Event Registration Capacity Controller
def event_registration_controller(capacity, total_registrations):
    
    confirmed = 0
    waitlisted = 0
    
    # Loop through registrations
    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    
    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")
event_registration_controller(100, 105)
```
