```python
#Problem Statement 1: Employee Performance Bonus Eligibility
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Temporary highest score
highest_score = None

#  Find employees with the highest score by comparison
for score in employees.values():
    if highest_score is None or score > highest_score:
        highest_score = score

# Getting dictionary of names of employee with the highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Display result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
#Problem Statement 2: Search Query Keyword Analysis
import string
query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

#  Remove punctuation 
for char in string.punctuation:
    query = query.replace(char, "")

# Split into words
words = query.split()

# Count frequency
frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Filter words searched more than once
result = {}

for word, count in frequency.items():
    if count > 1:
        result[word] = count

print(result)
```

```python
# Problem Statement 3: Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

# Loop through list with index
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:   # Check if even value
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)

 
```

```python
# Problem Statement 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Extract domain and count frequency
for email in emails:
    domain = email.split("@")[1]
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Calculate percentage usage
total_users = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
#nProblem Statement 5: Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average sales
total_sales = sum(sales)
average_sales = total_sales / len(sales)

# Calculate 30% above average threshold
threshold = average_sales * 1.3

# Detect spike days
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
# Problem 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count frequency of each user ID
id_count = {}

for user in user_ids:
    if user in id_count:
        id_count[user] += 1
    else:
        id_count[user] = 1

# Display only duplicates (count > 1)
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
