```python
#Problem 1: Social Media – Post Engagement Analyzer
#function to calculate the likes
def analyze_engagement(likes_list):
    total_likes = 0
    
    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like
    
    # Check post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    # Display output
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example usage
likes = [200, 300, 100, 400, 50]
analyze_engagement(likes)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker
def rainfall_checker(rainfall_list, required_level):
    
    total_rainfall = 0
    
    # Loop to calculate total rainfall
    for rain in rainfall_list:
        total_rainfall += rain
    
    # Calculate average rainfall
    average = total_rainfall / len(rainfall_list)
    
    # Check rainfall adequacy
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    # Display output
    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)


# Example usage
rainfall_data = [70, 75, 68, 80, 67]
rainfall_checker(rainfall_data, 70)
```

```python
# Problem 4: Social Media – Duplicate Account Detection
def detect_duplicate_accounts(usernames):
    
    # Convert list to set
    unique_users = set(usernames)
    
    # Check if duplicates exist
    if len(unique_users) < len(usernames):
        result = "Yes"
    else:
        result = "No"
    
    # Display output
    print("Duplicate Accounts Found:", result)


# Example usage
users = ["user1", "user2", "user3", "user1"]
detect_duplicate_accounts(users)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter
def filter_premium_crops(prices):
    
    premium_crops = []
    
    # Loop to filter premium prices
    for price in prices:
        if price > 2000: #if price>2000 it add to the list
            premium_crops.append(price)
    
    # Display output
    print("Premium Crops:", premium_crops)


# Example usage
crop_prices = [1800, 2500, 1500, 3200, 1900]
filter_premium_crops(crop_prices)
```

```python
# Problem 7: System Monitoring – Application Health Checker
def application_health_checker(error_count):
    
    # Check system health
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    # Display output
    print("Error Count:", error_count)
    print("System Status:", status)


# Example usage
application_health_checker(7)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker
def transaction_limit_checker(amount):
    
    daily_limit = 50000
    
    # Check transaction limit
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    
    # Display output
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


# Example usage
transaction_limit_checker(60000)
```

```python
#Problem 9: E-Learning – Student Attendance Eligibility System
def check_attendance_eligibility(attendance_list):
    
    total_classes = len(attendance_list)
    present_count = 0
    
    # Loop to count present days
    for status in attendance_list:
        if status == "P":
            present_count += 1
    
    # Calculate attendance percentage
    percentage = (present_count / total_classes) * 100
    
    # Check eligibility
    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    
    # Display output
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)


# Example usage
attendance = ["P", "P", "A", "P", "P", "P", "P", "A", "P", "P"]  # 8 out of 10 = 80%
check_attendance_eligibility(attendance)
```

```python
# Problem 10 : Smart Electricity Bill Analyzer
def electricity_bill_analyzer(units):
    
    bill = 0
    
    # Calculate bill based on slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Classify usage
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    
    return bill, usage


# Example usage
total_bill, usage_status = electricity_bill_analyzer(250)

print("Units Consumed:", 250)
print("Total Bill:", total_bill)
print("Usage Category:", usage_status)
```
