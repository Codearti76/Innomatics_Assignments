```python
#Problem 1: Mobile Recharge Validation System
def validate_recharge(amount, valid_plans):
    if amount < 50: #as minimum required value should be more than 50
        print("❌ Recharge amount must be at least ₹50.")
        return False
    elif amount not in valid_plans:
        print("❌ Invalid plan selected.")
        return False
    else:
        print("✅ Recharge successful!")
        return True


valid_plans = [199, 299, 399, 599]

while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))
        
        
        if validate_recharge(amount, valid_plans):
            print(f"₹{amount}")
            break   # Exit loop if valid
            
        print("Please try again.\n")
        
    except ValueError:
        print("❌ Please enter a valid numeric amount.\n")
```

```python
# Problem 2: Inventory Reorder Alert System
def check_inventory(products):
    for product in products:
        stock = products[product]
        
        if stock < 15:
            print(product, "→ Reorder Alert (Stock:", stock, ")")
        else:
            print(product, "→ Stock OK (Stock:", stock, ")")


#Dictionary of items
inventory = {
    "Rice": 10,
    "Wheat": 25,
    "Sugar": 12,
    "Oil": 18,
    "Peas" :15
}

check_inventory(inventory)

```

```python
# Problem 3: Student Result Processing System
def check_result(marks):
    total = 0

    for i in marks:
        total = total + i

    average = total / len(marks) 

    print("Average Marks:", average)

    if average >= 50:
        print("Pass")
    else:
        print("Fail")


# Example
marks = [40, 60, 50, 30, 60]
check_result(marks)
```

```python
# Problem 4: Cab Fare Estimator with Retry Option

def calculate_fare(distance, peak):
    fare = 50 + (12 * distance)

    if peak == "yes":
        fare = fare + (fare * 0.25)

    return fare


while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")

    total = calculate_fare(distance, peak) #store the return value

    print("Total Fare: ₹", total)

    choice = input("Do you want to try again? (yes/no): ")
    if choice != "yes":
        break
```

```python
#Problem 5: Employee Attendance Eligibility Checker

def check_eligibility(attendance):
    present = 0

    for i in attendance:
        if i == "P":
            present = present + 1

    percentage = (present /len(attendance)) *100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        print("Eligible")
    else:
        print("Not Eligible")


# Example list of attendance
attendance = ["P", "A", "P", "P", "A", "P", "P", "P", "A"]
check_eligibility(attendance)
```

```python
#Problem 6: Password Strength Checker
def check_password(password):
    if len(password) < 8:
        return "Weak Password"
    
    has_digit = False
    has_special = False
    
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True
    
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# Example Usage
print(check_password("abc12#"))
print(check_password("password@123"))
print(check_password("innomatics@26"))


```
